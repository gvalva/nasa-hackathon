```python
import pandas as pd

country_pollution_df = pd.read_csv('EPIdataAir.csv')

# dataset has daily totals per country so sort by date then group by country and take the first in the group
country_daily_covid_cases_df = pd.read_csv('owid-covid-data.csv') \
    .sort_values(['date'], ascending = False) \
    .groupby('iso_code') \
    .first() \
    .reset_index()

country_pollution_covid_cases_df = pd \
    .merge(country_pollution_df,country_daily_covid_cases_df, how = 'inner', left_on ='iso', right_on = 'iso_code')

country_pollution_covid_cases_df
                                           
                                           
                                           
```
